```python
# KMP
def calculate_LSP(pattern):
    LSP = [0]
    prefix_len = 0
    i = 1
    
    while i < len(pattern):
        # AA
        if pattern[prefix_len] == pattern[i]:
            prefix_len += 1
            LSP.append(prefix_len)
            i += 1
        else:
            ## AB
            if prefix_len == 0:
                LSP.append(0)
                i += 1
            ## AAC
            else:
                prefix_len = LSP[prefix_len - 1]
    return LSP
    
            

def KMP(main, pattern):
    LSP = calculate_LSP(pattern)
    print(f'LSP: {LSP}')
    j = 0
    i = 0
    while i < len(main):
        if pattern[j] == main[i]:
            if j == len(pattern)-1:
                return True
            j += 1
            i += 1
        elif j == 0:
            i += 1
        else:
            j = LSP[j-1]
    return False

main = "ABABABCAA"
pattern = "ABABC"
print(KMP(main, pattern))
```

```python
class Sorting:
    # Bubble Sort
    def bubble(self, nums):
        """
        TC: O(n^2)
        SC: O(1)
        Comparison: Yes
        Stable: Yes
        """
        n = len(nums)
        for i in range(n):
            for j in range(n - i - 1):
                if nums[j] > nums[j + 1]:
                    nums[j], nums[j + 1] = nums[j + 1], nums[j]
        return nums

    # Selection Sort
    def selection(self, nums):
        """
        TC: O(n^2)
        SC: O(1)
        Comparison: Yes
        Stable: No (e.g. [5,2,5,1])
        """
        for i in range(len(nums)):
            min_index = i
            for j in range(i + 1, len(nums)):
                if nums[j] < nums[min_index]:
                    min_index = j
            nums[i], nums[min_index] = nums[min_index], nums[i]
        return nums

    # Insertion Sort
    def insertion(self, nums):
        """
        TC: O(n^2), best O(n)
        SC: O(1)
        Comparison: Yes
        Stable: Yes
        """
        for i in range(1, len(nums)):
            key = nums[i]
            j = i - 1
            while j >= 0 and nums[j] > key:
                nums[j + 1] = nums[j]
                j -= 1
            nums[j + 1] = key   # 修正：应插入在 j+1
        return nums

    # Merge Sort
    def merge_sort(self, nums):
        """
        TC: O(n log n)
        SC: O(n)
        Comparison: Yes
        Stable: Yes
        Recursive: Yes
        """
        if len(nums) <= 1:
            return nums

        mid = len(nums) // 2
        left = self.merge_sort(nums[:mid])
        right = self.merge_sort(nums[mid:])

        return self._merge(left, right)

    def _merge(self, left, right):
        result = []
        i = j = 0
        while i < len(left) and j < len(right):  # 修正：要用 while
            if left[i] <= right[j]:
                result.append(left[i])
                i += 1
            else:
                result.append(right[j])
                j += 1
        result.extend(left[i:])
        result.extend(right[j:])
        return result

    # Quick Sort
    def quick_sort(self, nums):
        """
        TC: O(n log n) average, O(n^2) worst
        SC: O(log n) recursion stack
        Comparison: Yes
        Stable: No
        Recursive: Yes
        """
        if len(nums) <= 1:
            return nums
        pivot = nums[len(nums) // 2]
        left = [x for x in nums if x < pivot]
        mid = [x for x in nums if x == pivot]
        right = [x for x in nums if x > pivot]
        return self.quick_sort(left) + mid + self.quick_sort(right)
```
